```python
combine_restaurant_df = pd.concat([tourist_entertainment_restaurant_df, regular_restaurant_df, sightseeing_restaurant_df, resting_restaurant_df])
combine_restaurant_df
```

시설규모가 중요한데 Nan과 0이 공존하기 때문에 Nan을 0으로 채워줍니다.


```python
combine_restaurant_df['시설규모'] = combine_restaurant_df['시설규모'].fillna(0)
combine_restaurant_df
```

## 일차 정제 완료
우리가 더 해야할일은 아래와 같다.
- 식사가능인원을 만들어야한다.
- 식사 가능 인원은 전체 면적의 70% 이다.

```python
def calc_people(x):
    return (x / 100) * 70

combine_restaurant_df['식사가능인원'] = combine_restaurant_df['시설규모'].apply(calc_people)
combine_restaurant_df
```

```python
combine_restaurant_df.describe()
```

```python
combine_restaurant_df = combine_restaurant_df[combine_restaurant_df.식사가능인원 > 0]
combine_restaurant_df = combine_restaurant_df[combine_restaurant_df.시설규모 > 16] # 16.52 가아니다.
combine_restaurant_df
```
